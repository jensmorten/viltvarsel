```python
df['Dato'].count()
```

```python
df=df[df['År']==2025]
df=df[df['Art'].isin(['Elg', 'Hjort', 'Rådyr'])]  
df=df[df['vegkategori'].isin(['E','F','K'])]
```

```python
df=df[df['ÅDT, total']>100]
```

```python
df['Dato'].count()
```

```python
df=df.dropna()
```

```python
df["UTM33_øst_int"] = (
    df["UTM33 øst"]
    .astype(str)
    .str.split(",", n=1)
    .str[0].astype(int)
)
```

```python
df["UTM_nord_int"] = (
    df["UTM33 nord"]
    .astype(str)
    .str.split(",", n=1)
    .str[0].astype(int)
)
```

```python
df=df[['Vegobjekt_540_id', 'Art','ÅDT, total','Vegobjekt_540_lengde', 'UTM_nord_int', 'UTM33_øst_int']]
```

```python
df['UTM_nord_avg']=df.groupby(["Vegobjekt_540_id", "Art"])['UTM_nord_int'].mean()
```

```python
df=df[['Vegobjekt_540_id', 'Art','ÅDT, total','Vegobjekt_540_lengde']]###, 'UTM_nord_str', 'UTM33_øst_str']]
```

```python
df["antall_kollisjoner"] = (
    df
    .groupby(["Vegobjekt_540_id", "Art"])
    .transform("size")
)
```

```python
df=df[df["antall_kollisjoner"] >=3]
```

```python
df["frekvens"] = (
    df["antall_kollisjoner"] / (365*round(df["ÅDT, total"],0)+round(df["Vegobjekt_540_lengde"],0))
)
##kollisjon per meter-bil-år
```

```python
df=df.sort_values('frekvens', ascending=False)
```

```python
df=df.drop_duplicates()
```

```python
df.to_csv("frekvens.csv",encoding='utf-8', index=False)
```

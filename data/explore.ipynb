```python
yrkesrisiko = pd.DataFrame({
    "yrkesgruppe": [
        "Ledere",
        "Høyskoleyrker",
        "Kontoryrker",
        "Salgs- og serviceyrker",
        "Renholdere, hjelpearbeidere mv.",
        "Bønder, fiskere mv.",
        "Prosess- og maskinoperatører, transportarbeidere mv.",
        "Håndverkere",
    ],
    "ulykker_per_arsverk": [
        4.0 / 1000,
        6.5 / 1000,
        7.7 / 1000,
        13.9 / 1000,
        15.5 / 1000,
        19.2 / 1000,
        20.1 / 1000,
        20.5 / 1000,
    ],
    "intervall_lav": [
        3.0 / 1000,
        5.0 / 1000,
        7.0 / 1000,
        12.0 / 1000,
        14.5 / 1000,
        17.5 / 1000,
        19.6 / 1000,
        20.2 / 1000,
    ],
    "intervall_høg": [
        5.0 / 1000,
        7.0 /1000,
        12.0 / 1000,
        14.5 / 1000,
        17.5 / 1000,
        19.6 / 1000,
        20.2 / 1000,
        20.7 /1000
    ]
    
})
```

```python
def map_arsrisiko_til_yrke(arsrisiko, yrkesdf):
    """
    Mapper årsrisiko (per årsverk) til illustrativ yrkessammenlikning.

    Parametre
    ----------
    arsrisiko : float
        Risiko per årsverk (f.eks. 0.012 = 12 per 1000)
    yrkesdf : pandas.DataFrame
        Må innehalde kolonnane:
        - 'yrkesgruppe'
        - 'intervall_lav'
        - 'intervall_høg'

    Returnerer
    ----------
    str : tekst for illustrativ samanlikning
    """

    # Sorter for tryggleik (lav → høg risiko)
    yrkesdf = yrkesdf.sort_values("intervall_lav")

    # Lågare enn lågaste kjente yrke (Ledere)
    if arsrisiko < yrkesdf["intervall_lav"].min():
        return "Lavere enn kjente yrker"

    # Høgare enn høgaste kjente yrke (Håndverkere)
    if arsrisiko > yrkesdf["intervall_høg"].max():
        return "Høyere enn høyrisikoyrker"

    # Finn intervall som treff
    treff = yrkesdf[
        (yrkesdf["intervall_lav"] <= arsrisiko) &
        (yrkesdf["intervall_høg"] >= arsrisiko)
    ]

    if not treff.empty:
        # Dersom fleire treff (overlapp), vel nærmaste midtpunkt
        treff = treff.copy()
        treff["avstand"] = (treff["ulykker_per_arsverk"] - arsrisiko).abs()
        return treff.sort_values("avstand").iloc[0]["yrkesgruppe"]

    # Fallback (burde eigentleg ikkje skje)
    return "Ukjent risikonivå"
```

```python
df["sammenlikning_yrke"] = df["årsrisiko"].apply(
    lambda x: map_arsrisiko_til_yrke(x, yrkesrisiko)
)
```

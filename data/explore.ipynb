```python
import pandas as pd
```

```python
#df = pd.read_csv("Fallvilt_trdlag_2016-2026_vegobjekter.csv", sep=";", low_memory=False)
#df = pd.read_csv("Fallvilt_trdlag_2016-2026_veglenkesekvenslengde.csv",sep=";", low_memory=False)
df = pd.read_csv("Fallvilt_trdlag_2016-2026_adttotallengder.csv",sep=";", low_memory=False)
```

```python
df.columns
```

```python
df.head(5)
```

```python
df['Dato'].count()
```

```python
df=df[df['År']==2025]
df=df[df['Art'].isin(['Elg', 'Hjort', 'Rådyr'])]  
df=df[df['vegkategori'].isin(['E','F','K'])]
```

```python
df=df[df['ÅDT, total']>100]
```

```python
df['Dato'].count()
```

```python
df["vegstrekning_kort"] = (
    df["vegsystemreferanse.kortform"]
    .str.split()
    .str[:2]
    .str.join("_")
)
```

```python
df=df.dropna()
```

```python
df=df[['Vegobjekt_540_id', 'Art', "ÅDT, total", "Vegobjekt_540_lengde"]]
```

```python
df.sort_values('Vegobjekt_540_id', ascending=False)
```

```python
df["antall_kollisjoner"] = (
    df
    .groupby(["Vegobjekt_540_id", "Art"])
    .transform("size")
)
```

```python
df=df[df["antall_kollisjoner"] >=3]
```

```python
df["frekvens"] = (
    df["antall_kollisjoner"] / (365*round(df["ÅDT, total"],0)+round(df["Vegobjekt_540_lengde"],0))
)
##kollisjon per meter-bil-år
```

```python
df=df.sort_values('frekvens', ascending=False)
```

```python
df=df.drop_duplicates()
```

```python
#df = pd.read_csv("Fallvilt_trdlag_2016-2026_vegobjekter.csv", sep=";", low_memory=False)
df = pd.read_csv("Fallvilt_trdlag_2016-2026_veglenkesekvenslengde.csv",sep=";", low_memory=False)
```

```python
df.columns
```

```python
df.head(5)
```

```python
df=df[df['År']==2025]
df = df[df['Art'].isin(['Elg', 'Hjort', 'Rådyr'])]  
```

```python
df["vegstrekning_kort"] = (
    df["vegsystemreferanse.kortform"]
    .str.split()
    .str[:2]
    .str.join("_")
)
```

```python
df=df.dropna()
```

```python
df=df[['vegstrekning_kort','veglenkesekvensid', 'Art', "ÅDT, total"]]
```

```python
df.sort_values('veglenkesekvensid', ascending=False)
```

```python
df=df[['vegstrekning_kort', 'veglenkesekvensid', 'Art', "ÅDT, total"]]
df=df.dropna()
```

```python
df.sort_values('veglenkesekvensid', ascending=False)
```

```python
df = (
    df
    .groupby(["veglenkesekvensid", "Art"])
    .agg(
        antall_kollisjoner=("Art", "size"),
        ÅDT_total=("ÅDT, total", "first"),
        Veglenkesekvenslengde = ("Veglenkesekvenslengde", "first")
    )
    .reset_index()
)

df["frekvens"] = (
    df["antall_kollisjoner"] / (df["ÅDT_total"]+df["Veglenkesekvenslengde"])
)
```
